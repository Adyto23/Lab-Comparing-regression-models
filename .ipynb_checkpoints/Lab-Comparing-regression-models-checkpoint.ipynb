```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
##Concatenate Numerical and Categorical dataframes into one dataframe called data.## was concated in the previous lab 


customer_df = pd.read_csv('concat.csv')
display(customer_df.shape)
customer_df.head()
customer_df=customer_df.drop(['year_year'], axis=1)
```

```python
customer_df.describe()
```

```python
#separate the features from the labels
y = customer_df['total_claim_amount']
X = customer_df.drop(['total_claim_amount'], axis=1)
```

```python
## In this final lab, we will model our data. Import sklearn train_test_split and separate the data.


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train.describe()
```

```python
numericals_train = X_train.select_dtypes(np.number)
numericals_test = X_test.select_dtypes(np.number)
numericals_train.head()

##split the X_train and X_test to numeric
```

```python
from sklearn.preprocessing import StandardScaler

transformer = StandardScaler().fit(numericals_train)
numericals_train_standardized = transformer.transform(numericals_train)
numericals_test_standardized = transformer.transform(numericals_test)

## StandardScaler to numeric , could use the minmax instead 
```

```python
pd.DataFrame(numericals_train_standardized).head()
```

```python
categoricals_train= X_train.select_dtypes(object)
categoricals_test= X_test.select_dtypes(object)
categoricals_train.head()

##split the X_train and X_test to to categorical
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder= OrdinalEncoder()

ordinal_train_cat=ordinal_encoder.fit_transform(categoricals_train[["coverage","employmentstatus","location_code","vehicle_size"]])
ordinal_test_cat=ordinal_encoder.fit_transform(categoricals_test[["coverage","employmentstatus","location_code","vehicle_size"]])

## ordinal_train is set on "coverage","employmentstatus","location_code","vehicle_size"
```

```python
categoricals_train=categoricals_train.drop(['coverage', 'employmentstatus','location_code','vehicle_size'], axis=1)
categoricals_test=categoricals_test.drop(['coverage', 'employmentstatus','location_code','vehicle_size'], axis=1)

## "coverage","employmentstatus","location_code","vehicle_size" was dropped from categoricals_train and test
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='error', drop='first').fit(categoricals_train)
categoricals_train_encoded = encoder.transform(categoricals_train).toarray()
categoricals_test_encoded = encoder.transform(categoricals_test).toarray()

## OneHotEncoder was made on the rest of the columns 
```

```python
encoder.categories_
```

```python
categoricals_test_encoded
```

```python
numericals_train_standardized
categoricals_train_encoded
```

```python
X_train_processed = np.concatenate((numericals_train_standardized,categoricals_train_encoded,ordinal_train_cat),axis=1)
X_test_processed = np.concatenate((numericals_test_standardized,categoricals_test_encoded,ordinal_test_cat),axis=1)

## concat back all the columns after standerized and encoding
```

```python
pd.DataFrame(X_train_processed)
```

```python
from sklearn import linear_model

LR = linear_model.LinearRegression()
LR.fit(X_train_processed,y_train)

## LinearRegression part of lab
```

```python
score = LR.score(X_test_processed, y_test)
score
```

```python
score = LR.score(X_train_processed, y_train)
score
```

```python
from sklearn.metrics import r2_score
predictions = LR.predict(X_test_processed)
r2_score(y_test, predictions)
```
